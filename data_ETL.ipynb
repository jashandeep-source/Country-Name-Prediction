```python
import pandas as pd
import numpy as np
```

```python
#read data
cities_data = pd.read_json('Data/CitiesData.json.gz')
```

```python
cities_data
```

```python
#get needed columns
cities_data = cities_data[['geoname_id', 'ascii_name', 'feature_code', 'cou_name_en', 'population', 'elevation', 'dem', 'coordinates', 'timezone','modification_date']]
```

```python
#rename columns
cities_data = cities_data.rename(columns={'cou_name_en': 'Country Name', 'ascii_name': 'area_name'})
```

```python
cities_data['modification_date'] = pd.to_datetime(cities_data['modification_date'], format='mixed',utc=True )
```

```python
#get year from modification date
cities_data['Year'] = cities_data['modification_date'].dt.year
```

```python
#read tourist_data for arrivals
tourist_data_arrivals = pd.read_csv('Data/InternationalTourism/numberOfArrivals.csv')
```

```python
tourist_data_arrivals = tourist_data_arrivals.drop(columns= ["Indicator Name", "Indicator Code"])
tourist_data_arrivals
```

```python
#convert data into needed format
tourist_data_arrivals = pd.melt(tourist_data_arrivals,id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="Number of Arrivals")
```

```python
tourist_data_arrivals 
```

```python
tourist_data_arrivals.dropna(subset=['Number of Arrivals'], inplace=True)
```

```python
tourist_data_departures = pd.read_csv('Data/InternationalTourism/numberOfDepartures.csv')
```

```python
tourist_data_departures
```

```python
#read tourist_data for departures
tourist_data_departures = tourist_data_departures.drop(columns= ["Indicator Name", "Indicator Code"])
```

```python
#convert data into needed format
tourist_data_departures = pd.melt(tourist_data_departures,id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="Number of Departures")
```

```python
tourist_data_departures.dropna(subset=['Number of Departures'], inplace=True)
```

```python
tourist_data_departures
```

```python
#read tourist_data for expenditures
tourist_data_expenditures = pd.read_csv('Data/InternationalTourism/expenditures.csv')
```

```python
tourist_data_expenditures
```

```python
tourist_data_expenditures = tourist_data_expenditures.drop(columns= ["Indicator Name", "Indicator Code"])
```

```python
#convert data into needed format
tourist_data_expenditures = pd.melt(tourist_data_expenditures,id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="Expenditures")
```

```python
tourist_data_expenditures.dropna(subset=['Expenditures'], inplace=True)
```

```python
tourist_data_expenditures
```

```python
#merge data
tourist_data = tourist_data_arrivals.merge(tourist_data_departures,on=['Country Name','Country Code', 'Year'])
```

```python
#merge data
tourist_data = tourist_data.merge(tourist_data_expenditures,on=['Country Name','Country Code', 'Year'])
```

```python
tourist_data
```

```python
#read unemployment_data
unemployment_data = pd.read_csv('Data/unemployment analysis.csv')
```

```python
unemployment_data
```

```python
#convert data into needed format
unemployment_data = pd.melt(unemployment_data,id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="Unemployment rate")
unemployment_data
```

```python
#merge data
data_merge = tourist_data.merge(unemployment_data,on=['Country Name','Country Code', 'Year'])
```

```python
data_merge['Year']=data_merge['Year'].astype(int) 
```

```python
#merge data
data = cities_data.merge(data_merge, left_on=['Country Name', 'Year'], right_on=['Country Name', 'Year'], how='inner')
```

```python
#group data by year to get count of countries per year
data_stats = data.groupby(["Year"]).agg({"Country Name":"count"})
data_stats
```

```python
#seems like year >=2010 is useful data 
data = data[data['Year'] >= 2010]
```

```python
#drop data where elevation has no values
data.dropna(subset=["elevation"],inplace=True)
```

```python
#save merged data to an input file
data.to_csv('Data/merged_data.csv', index=False)
```
